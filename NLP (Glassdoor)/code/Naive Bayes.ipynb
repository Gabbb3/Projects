```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf


df = pd.read_csv("reviews_final_modelled.csv")

df.head()
```

```python
print(df['overall_review'].count(),'\n\n',df['overall_rating'].value_counts(),'\n\n',
      df['overall_review'].count()==df['overall_rating'].count())
```

```python
rating = []
for i in df['overall_rating']:
        if i > 3 : rating.append(1)
        elif i > 2:  rating.append(0)
        else:       rating.append(-1)

df['overall_rating2'] = rating
    
df.head()
```

```python
print(df['overall_review'].count(),'\n\n',df['overall_rating2'].value_counts(),'\n\n',
      df['overall_review'].count()==df['overall_rating2'].count())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split,StratifiedKFold
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import matthews_corrcoef
```

```python
train_data = df
train_data.shape
```

```python
vectorizer = CountVectorizer(analyzer="word", 
                             preprocessor=None, 
                             stop_words=None, 
                             ngram_range=(1,1),
                            max_features=5000)
```

```python
train_data_features = vectorizer.fit_transform(df.cleaned_review)
train_data_features = train_data_features.toarray()
```

```python
X = train_data_features
y = train_data.overall_rating2
```

```python
print (X.shape) 
print (y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=123)
```

```python
naive = MultinomialNB()
classifier = naive.fit(X_train,y_train)
predict = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(predict,y_test)

print(cm)
print()
print(classification_report(predict,y_test))
print("MCC-Score:", matthews_corrcoef(y_test,predict))
```

```python
accuracy = cm.trace()/cm.sum()
print(accuracy)
```
